```python
# Install required packages
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from urllib.parse import urljoin
import pandas as pd

# Set up Chrome options for Colab
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Initialize the driver
driver = webdriver.Chrome(options=chrome_options)

def extract_urls_from_page(page_number):
    """Extract all law URLs from a specific page"""
    url = f"https://placng.org/lawsofnigeria/?page={page_number}"

    try:
        print(f"Processing page {page_number}...")
        driver.get(url)

        # Wait for the page to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )

        # Find all links on the page
        all_links = driver.find_elements(By.TAG_NAME, "a")

        law_urls = []
        for link in all_links:
            href = link.get_attribute("href")
            if href:
                # Filter for the specific URL patterns we want
                if ("https://placng.org/lawsofnigeria/laws/" in href or
                    "https://placng.org/lawsofnigeria/view2.php?sn=" in href):
                    law_urls.append(href)

        return list(set(law_urls))  # Remove duplicates

    except Exception as e:
        print(f"Error processing page {page_number}: {e}")
        return []

def main():
    all_law_urls = []
    total_pages = 69

    for page_num in range(1, total_pages + 1):
        page_urls = extract_urls_from_page(page_num)
        all_law_urls.extend(page_urls)

        # Small delay to be respectful to the server
        time.sleep(1)

        # Progress update
        print(f"Page {page_num}: Found {len(page_urls)} law URLs")

    # Remove duplicates and sort
    unique_urls = sorted(list(set(all_law_urls)))

    # Print summary
    print(f"\n=== EXTRACTION COMPLETE ===")
    print(f"Total pages processed: {total_pages}")
    print(f"Total unique law URLs found: {len(unique_urls)}")

    # Display first few URLs as sample
    print(f"\nSample URLs (first 10):")
    for i, url in enumerate(unique_urls[:10]):
        print(f"{i+1}. {url}")

    # Save to CSV file
    df = pd.DataFrame(unique_urls, columns=['Law_URLs'])
    df.to_csv('nigeria_law_urls.csv', index=False)
    print(f"\nAll URLs saved to 'nigeria_law_urls.csv'")

    return unique_urls

# Run the extraction
if __name__ == "__main__":
    try:
        law_urls = main()
    finally:
        # Always close the driver
        driver.quit()
```

```python
# Install required packages
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
!pip install selenium trafilatura requests pandas tqdm

import os
import time
import json
import requests
import pandas as pd
from urllib.parse import urlparse, unquote
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, WebDriverException
import trafilatura
from tqdm import tqdm
from datetime import datetime

# Create main output directory with timestamp
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
main_output_dir = f"nigeria_laws_data_{timestamp}"
os.makedirs(main_output_dir, exist_ok=True)

# Create subdirectories within the main folder
pdfs_dir = os.path.join(main_output_dir, 'pdfs')
texts_dir = os.path.join(main_output_dir, 'texts')
os.makedirs(pdfs_dir, exist_ok=True)
os.makedirs(texts_dir, exist_ok=True)

print(f"üìÅ Main output directory: {main_output_dir}")
print(f"üìÑ PDFs will be saved in: {pdfs_dir}")
print(f"üìù Texts will be saved in: {texts_dir}")

# Set up Chrome options for Colab
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)

# Initialize the driver
driver = webdriver.Chrome(options=chrome_options)
driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

def setup_requests_session():
    """Set up requests session with headers"""
    session = requests.Session()
    session.headers.update({
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    })
    return session

def is_pdf_url(url):
    """Check if URL points to a PDF file"""
    parsed = urlparse(url)
    path = parsed.path.lower()
    return path.endswith('.pdf') or 'pdf' in parsed.query.lower()

def get_filename_from_url(url, content_type=None):
    """Extract filename from URL"""
    parsed = urlparse(url)
    path = unquote(parsed.path)

    if is_pdf_url(url):
        # For PDF URLs, try to get filename from path
        filename = os.path.basename(path)
        if not filename or '.' not in filename:
            filename = f"document_{hash(url) % 10000}.pdf"
        elif not filename.endswith('.pdf'):
            filename += '.pdf'
    else:
        # For HTML pages, create a descriptive filename
        filename = path.replace('/', '_').replace('?', '_').replace('&', '_')
        if not filename or filename == '_':
            filename = f"page_{hash(url) % 10000}"
        if not filename.endswith('.txt'):
            filename += '.txt'

    # Clean filename
    filename = "".join(c for c in filename if c.isalnum() or c in "._-")
    return filename

def download_pdf(session, url, max_retries=3):
    """Download PDF file"""
    for attempt in range(max_retries):
        try:
            response = session.get(url, stream=True, timeout=30)
            response.raise_for_status()

            # Check if content is actually PDF
            content_type = response.headers.get('content-type', '').lower()
            if 'pdf' not in content_type and not is_pdf_url(url):
                return False, "Not a PDF file"

            filename = get_filename_from_url(url)
            filepath = os.path.join(pdfs_dir, filename)

            with open(filepath, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)

            return True, filepath

        except Exception as e:
            if attempt < max_retries - 1:
                time.sleep(2)
            else:
                return False, f"Failed after {max_retries} attempts: {str(e)}"

def extract_page_text_with_selenium(url, max_retries=2):
    """Extract text from HTML page using Selenium (for JavaScript-heavy pages)"""
    for attempt in range(max_retries):
        try:
            driver.get(url)
            WebDriverWait(driver, 15).until(
                EC.presence_of_element_located((By.TAG_NAME, "body"))
            )

            # Get page source and extract with trafilatura
            html_content = driver.page_source
            text = trafilatura.extract(html_content)

            if text and len(text.strip()) > 100:  # Ensure meaningful content
                return True, text
            else:
                # Fallback: try to get text directly from body
                body_text = driver.find_element(By.TAG_NAME, "body").text
                if body_text and len(body_text.strip()) > 100:
                    return True, body_text
                else:
                    return False, "No meaningful text content found"

        except Exception as e:
            if attempt < max_retries - 1:
                time.sleep(2)
            else:
                return False, f"Failed to extract text: {str(e)}"

def extract_page_text_with_requests(session, url, max_retries=2):
    """Extract text from HTML page using requests + trafilatura"""
    for attempt in range(max_retries):
        try:
            response = session.get(url, timeout=30)
            response.raise_for_status()

            text = trafilatura.extract(response.text)
            if text and len(text.strip()) > 100:
                return True, text
            else:
                return False, "No meaningful text content found"

        except Exception as e:
            if attempt < max_retries - 1:
                time.sleep(1)
            else:
                return False, f"Failed to extract text: {str(e)}"

def process_url(session, url, index, total):
    """Process a single URL - download PDF or extract text"""
    result = {
        'url': url,
        'index': index,
        'filename': None,
        'filepath': None,
        'content_type': None,
        'success': False,
        'error': None,
        'timestamp': datetime.now().isoformat()
    }

    try:
        print(f"[{index}/{total}] Processing: {url}")

        if is_pdf_url(url):
            result['content_type'] = 'pdf'
            success, message = download_pdf(session, url)

            if success:
                result['success'] = True
                result['filepath'] = message
                result['filename'] = os.path.basename(message)
                print(f"    ‚úÖ PDF downloaded: {result['filename']}")
            else:
                result['error'] = message
                print(f"    ‚ùå PDF download failed: {message}")

        else:
            result['content_type'] = 'html'
            # First try with requests (faster)
            success, text_or_error = extract_page_text_with_requests(session, url)

            if not success:
                # Fallback to Selenium if requests fails
                success, text_or_error = extract_page_text_with_selenium(url)

            if success:
                result['success'] = True
                filename = get_filename_from_url(url)
                filepath = os.path.join(texts_dir, filename)

                # Save extracted text
                with open(filepath, 'w', encoding='utf-8') as f:
                    f.write(text_or_error)

                result['filepath'] = filepath
                result['filename'] = filename
                print(f"    ‚úÖ Text extracted: {filename} ({len(text_or_error)} characters)")
            else:
                result['error'] = text_or_error
                print(f"    ‚ùå Text extraction failed: {text_or_error}")

    except Exception as e:
        result['error'] = f"Unexpected error: {str(e)}"
        print(f"    ‚ùå Unexpected error: {str(e)}")

    return result

def load_urls_from_csv(csv_file):
    """Load URLs from CSV file"""
    try:
        df = pd.read_csv(csv_file)
        # Try common column names
        possible_columns = ['Law_URLs', 'url', 'URL', 'link', 'Link']
        for col in possible_columns:
            if col in df.columns:
                urls = df[col].dropna().unique().tolist()
                print(f"Found {len(urls)} URLs in column '{col}'")
                return urls

        # If no common column names found, use first column
        urls = df.iloc[:, 0].dropna().unique().tolist()
        print(f"Found {len(urls)} URLs in first column")
        return urls

    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return []

def load_urls_from_json(json_file):
    """Load URLs from JSON file (from previous extraction)"""
    try:
        with open(json_file, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # Handle different JSON formats
        if isinstance(data, list):
            # Simple array format
            urls = data
        elif 'urls' in data:
            # Our comprehensive format
            urls = data['urls']
        elif 'results' in data and 'all_urls' in data['results']:
            # Alternative format
            urls = data['results']['all_urls']
        else:
            # Try to find URLs in any list in the data
            urls = []
            for key, value in data.items():
                if isinstance(value, list) and len(value) > 0 and isinstance(value[0], str) and 'http' in value[0]:
                    urls.extend(value)

        print(f"Found {len(urls)} URLs in JSON file")
        return list(set(urls))  # Remove duplicates

    except Exception as e:
        print(f"Error reading JSON file: {e}")
        return []

def create_readme_file():
    return "none"
def main():
    # Try to load URLs from different sources
    urls = []

    # Try JSON first (from our previous extraction)
    json_file = 'nigeria_law_urls.json'
    if os.path.exists(json_file):
        urls = load_urls_from_json(json_file)

    # If no URLs from JSON, try CSV
    if not urls:
        csv_file = 'nigeria_law_urls.csv'
        if os.path.exists(csv_file):
            urls = load_urls_from_csv(csv_file)

    if not urls:
        print("‚ùå No URLs found. Please make sure you have either:")
        print("   - nigeria_law_urls.json (from previous extraction)")
        print("   - nigeria_law_urls.csv (CSV file with URLs)")
        return

    print(f"üöÄ Starting processing of {len(urls)} URLs...")

    # Set up session
    session = setup_requests_session()

    # Process URLs
    results = []
    successful_downloads = 0

    for i, url in enumerate(tqdm(urls, desc="Processing URLs")):
        result = process_url(session, url, i + 1, len(urls))
        results.append(result)

        if result['success']:
            successful_downloads += 1

        # Respectful delay between requests
        time.sleep(1)

    # Generate summary report
    pdf_count = sum(1 for r in results if r['content_type'] == 'pdf' and r['success'])
    html_count = sum(1 for r in results if r['content_type'] == 'html' and r['success'])
    failed_count = len(urls) - successful_downloads

    print(f"\n{'='*60}")
    print("üìä PROCESSING COMPLETE")
    print(f"{'='*60}")
    print(f"Total URLs processed: {len(urls)}")
    print(f"Successful downloads: {successful_downloads}")
    print(f"  - PDF files: {pdf_count}")
    print(f"  - Text extractions: {html_count}")
    print(f"Failed: {failed_count}")
    print(f"\nüìÅ All files saved in: {main_output_dir}")

    # Save processing results to JSON in the main directory
    results_data = {
        'metadata': {
            'processing_date': datetime.now().isoformat(),
            'main_output_directory': main_output_dir,
            'pdfs_directory': pdfs_dir,
            'texts_directory': texts_dir,
            'total_urls': len(urls),
            'successful_downloads': successful_downloads,
            'failed_downloads': failed_count,
            'pdf_downloads': pdf_count,
            'text_extractions': html_count
        },
        'results': results
    }

    results_json_path = os.path.join(main_output_dir, 'processing_results.json')
    with open(results_json_path, 'w', encoding='utf-8') as f:
        json.dump(results_data, f, indent=2, ensure_ascii=False)

    # Create README file
    readme_path = create_readme_file()

    print(f"\nüíæ Files created:")
    print(f"   üìÑ {results_json_path}")
    print(f"   üìñ {readme_path}")
    print(f"   üìÅ {pdfs_dir} ({pdf_count} PDF files)")
    print(f"   üìÅ {texts_dir} ({html_count} text files)")

    # Show failed URLs for debugging
    failed_urls = [r for r in results if not r['success']]
    if failed_urls:
        print(f"\n‚ùå Failed URLs ({len(failed_urls)}):")
        for result in failed_urls[:5]:  # Show first 5 failures
            print(f"   - {result['url']}: {result['error']}")

    # Display directory tree
    print(f"\nüå≥ Directory Structure:")
    print(f"{main_output_dir}/")
    print(f"‚îú‚îÄ‚îÄ pdfs/")
    print(f"‚îÇ   ‚îî‚îÄ‚îÄ [PDF files]")
    print(f"‚îú‚îÄ‚îÄ texts/")
    print(f"‚îÇ   ‚îî‚îÄ‚îÄ [Text files]")
    print(f"‚îú‚îÄ‚îÄ processing_results.json")
    print(f"‚îî‚îÄ‚îÄ README.md")

    return results_data

# Run the processing
if __name__ == "__main__":
    try:
        results = main()
    except KeyboardInterrupt:
        print("\n‚ö†Ô∏è Processing interrupted by user")
    except Exception as e:
        print(f"\n‚ùå Script failed with error: {e}")
    finally:
        # Always close the driver
        driver.quit()
        print(f"\nüéâ All data organized in: {main_output_dir}")
```

```python
# Install required packages
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
!pip install selenium pandas tqdm

import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from tqdm import tqdm
from datetime import datetime

# Set up Chrome options for Colab
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)

# Initialize the driver
driver = webdriver.Chrome(options=chrome_options)
driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

def extract_links_with_text_from_page(page_number):
    """Extract URLs and their link text for links matching the two patterns"""
    try:
        url = f"https://placng.org/lawsofnigeria/?page={page_number}"
        print(f"Processing page {page_number}...")
        driver.get(url)

        # Wait for page to load
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )

        links_data = []

        # Find all <a> tags on the page
        all_links = driver.find_elements(By.TAG_NAME, "a")

        for link in all_links:
            href = link.get_attribute('href')
            if href:
                # Check if href matches our two target patterns
                if ("https://placng.org/lawsofnigeria/laws/" in href or
                    "https://placng.org/lawsofnigeria/view2.php?sn=" in href):

                    link_text = link.text.strip()

                    # If no visible text, try other attributes
                    if not link_text:
                        link_text = link.get_attribute('title') or link.get_attribute('innerText') or ''
                        link_text = link_text.strip()

                    links_data.append({
                        'url': href,
                        'link_text': link_text
                    })

        print(f"Page {page_number}: Found {len(links_data)} matching links")
        return links_data

    except Exception as e:
        print(f"Error processing page {page_number}: {e}")
        return []

def create_enhanced_csv():
    """Create CSV with url and link_text columns for all matching links"""
    all_links_data = []
    total_pages = 69

    print(f"üéØ Extracting links matching patterns:")
    print(f"  1. https://placng.org/lawsofnigeria/laws/*")
    print(f"  2. https://placng.org/lawsofnigeria/view2.php?sn=*")
    print(f"üìÑ Processing {total_pages} pages...")

    for page_num in tqdm(range(1, total_pages + 1), desc="Extracting links"):
        page_links = extract_links_with_text_from_page(page_num)
        all_links_data.extend(page_links)
        time.sleep(0.5)  # Respectful delay

    # Remove duplicates while preserving order
    seen_urls = set()
    unique_links_data = []

    for link in all_links_data:
        if link['url'] not in seen_urls:
            seen_urls.add(link['url'])
            unique_links_data.append(link)

    # Create DataFrame
    df = pd.DataFrame(unique_links_data)

    # Add file type classification
    df['file_type'] = df['url'].apply(
        lambda x: 'PDF' if '/laws/' in x else 'HTML'
    )

    # Reorder columns
    df = df[['url', 'link_text', 'file_type']]

    # Save to CSV
    output_filename = "nigeria_laws_urls_with_link_text.csv"
    df.to_csv(output_filename, index=False)

    print(f"\n‚úÖ CSV created: {output_filename}")
    print(f"üìä Total unique links: {len(df)}")
    print(f"üìÑ PDF links (/laws/): {len(df[df['file_type'] == 'PDF'])}")
    print(f"üåê HTML links (view2.php): {len(df[df['file_type'] == 'HTML'])}")
    print(f"üìù Links with non-empty text: {len(df[df['link_text'] != ''])}")

    # Display sample
    print(f"\nüîç Sample data:")
    print(df.head(10).to_string(index=False))

    return df

# Run the extraction
try:
    df = create_enhanced_csv()
    print(f"\nüéâ Extraction complete! File saved as 'nigeria_laws_urls_with_link_text.csv'")
finally:
    driver.quit()
```
